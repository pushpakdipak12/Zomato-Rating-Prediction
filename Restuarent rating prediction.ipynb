```python
zomato=zomato.drop('name',axis=1)
```

```python
zomato['cuisines'].value_counts()
```

```python
zomato['cuisines_split'] = zomato['cuisines'].str.strip().str.split(', ')

# Create a set of unique cuisines
unique_cuisines = set()
zomato['cuisines_split'].apply(lambda x: unique_cuisines.update(x))

# Convert unique_cuisines set to a list for easy iteration
unique_cuisines = list(unique_cuisines)

# Create binary columns for each unique cuisine
for cuisine in unique_cuisines:
    # Create a binary column for each cuisine
    zomato[cuisine] = zomato['cuisines'].str.contains(cuisine).astype(int)
```
